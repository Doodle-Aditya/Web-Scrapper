```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
final = pd.DataFrame()
for j in range(1,11):
    url = f'https://quotes.toscrape.com/page/{j}/'
    webpage = requests.get(url).text

    soup = BeautifulSoup(webpage, 'lxml')
    quotes = soup.find_all('div',class_='quote')
    name = []
    quote = []
    tag =[]
    for i in quotes:
        name.append(i.find('small', class_='author').text)
        quote.append(i.find('span',class_='text').text)
        tags = i.find('a',class_='tag')
        if tags:
            tag.append(tags.text)
        else:
            tag.append('')
    d = {'name':name,'quote':quote,'category':tag}
    df = pd.DataFrame(d)

    final = pd.concat([final,df] , ignore_index=True)
```
